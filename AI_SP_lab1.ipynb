# Q1.)Load the recorded speech file into your python workspace. Once loaded, plot the graph for the speech signal.You may use the below code from librosa asa reference.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

y, sr = librosa.load("/content/lab1.wav")
librosa.display.waveshow(y)
plt.title('Speech Signal')
```

# Q2.) Observe the length & magnitude range of the signalfrom theplot. Observing the plot, relate tothe spoken words and silence between the words.

```python
duration = len(y) / sr  # Duration in seconds
print(f"Duration of the audio: {duration:.2f} seconds")
print(f"Magnitude range: {y.min()} to {y.max()}")
```

# Q3.) Take a small segment of the signal and play it.

```python
import IPython.display as ipd

start_time = 3
end_time = 4

# Extract the segment
segment = y[int(start_time * sr):int(end_time * sr)]

# Display the waveform
librosa.display.waveshow(segment)
plt.title('Speech Signal Segment')

# Play the segment
ipd.Audio(segment, rate=sr)
```
